```python
df_copy['Size']
```

```python
df_copy["Size"].unique() # i see some value with M or k so we have to convert this into int
```

```python
# converting the k and M into the blank space and after that convert all the
# things into the float
def convert_size_to_mb(size):
    if isinstance(size, float):
        return size
    if 'M' in size:
        return float(size.replace('M', '')) # replace the M with empty space
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024
    return np.nan

df_copy['Size'] = df_copy['Size'].apply(convert_size_to_mb)
```
